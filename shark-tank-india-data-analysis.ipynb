# New column of comparision between ask_equity and deal_equity--

```python
data['difference_equity'] = data.apply(lambda x: x['deal_equity'] - x['ask_equity'], axis = 1)
```

```python
data[["ask_equity","deal_equity","difference_equity"]]
```

# New column of comparision between ask_valuation and deal_valuation--

```python
#diference_valuation (deal_valuation- (asked_valuation))
data['difference_valuation'] = data.apply(lambda x: x['deal_valuation'] - x['ask_valuation'], axis = 1)
data[["ask_valuation","deal_valuation","difference_valuation"]]
```

# New column of comparision between ask_amount and deal_amount--

```python
data['difference_amount'] = data.apply(lambda x: x['deal_amount'] - x['pitcher_ask_amount'], axis = 1)
data[["pitcher_ask_amount","deal_amount","difference_amount"]]
```
